```python
#4) How does family income and parental education level impact the student's exam scores?


family_income = data['Family_Income']
parental_education = data['Parental_Involvement']
exam_scores = data['Exam_Score']
df = pd.DataFrame({'Family_Income': family_income, 'Parental_Involvement': parental_education, 'Exam_Score': exam_scores})
income_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
education_mapping = {'Low': 1, 'Medium': 2, 'High': 3}

df['Family_Income'] = df['Family_Income'].map(income_mapping)
df['Parental_Involvement'] = df['Parental_Involvement'].map(education_mapping)
corr = df.corr()
corr
```
