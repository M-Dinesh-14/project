1) How do confirmed cases and fatalities vary over time for different regions?


```python
data = pd.read_csv('/content/covid-19.csv')
region_data = data.groupby(['Date', 'Country/Region'])[['ConfirmedCases', 'Fatalities']].sum().reset_index()
```

```python
region_data
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(5,5))
for region in region_data['Country/Region'].unique():
    region_subset = region_data[region_data['Country/Region'] == region]
    plt.plot(region_subset['Date'], region_subset['ConfirmedCases'], label=f'{region} ConfirmedCases')
    plt.plot(region_subset['Date'], region_subset['Fatalities'], label=f'{region} Fatalities')

plt.title("Confirmed Cases and Fatalities Over Time by Region")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

2) Which countries or regions have the highest and lowest total confirmed cases and fatalities?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('/content/covid-19.csv')
data.head()

country_data = data.groupby('Country/Region')[['ConfirmedCases', 'Fatalities']].sum().reset_index()


country_data_sorted_cases = country_data.sort_values(by='ConfirmedCases', ascending=False)
country_data_sorted_fatalities = country_data.sort_values(by='Fatalities', ascending=False)
highest_cases = country_data_sorted_cases.head(1)
lowest_cases = country_data_sorted_cases.tail(1)
print("Region with the highest confirmed cases:")
print(highest_cases)

print("Region with the lowest confirmed cases:")
print(lowest_cases)
highest_fatalities = country_data_sorted_fatalities.head(1)
lowest_fatalities = country_data_sorted_fatalities.tail(1)


print("Region with the highest fatalities:")
print(highest_fatalities)

print("Region with the lowest fatalities:")
print(lowest_fatalities)
```

3) Is there a correlation between temperature (temp) and the number of confirmed cases or fatalities?


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot style
sns.set(style="whitegrid")

# Plotting temperature vs confirmed cases
plt.figure(figsize=(5,5))
sns.scatterplot(x='temp', y='ConfirmedCases', data=data, color='blue')
plt.title("Temperature vs Confirmed Cases")
plt.xlabel("Temperature (°C)")
plt.ylabel("ConfirmedCases")
plt.show()
```
